```python
import pandas as pd
df = pd.read_csv('/Users/darianaumenko/Downloads/habr.csv')
df
```

```python
df.drop(columns = ['status', 'id'], inplace= True)
df.drop(columns = 'time', inplace= True)
```

```python
import ast

def safe_parse_list(x):
    if isinstance(x, list):
        return x
    try:
        return ast.literal_eval(x)
    except:
        return []
df["keywords"] = df["keywords"].apply(safe_parse_list)
df["hubs"] = df["hubs"].apply(safe_parse_list)
print(df["keywords"].iloc[0], type(df["keywords"].iloc[0]))
print(df["hubs"].iloc[0], type(df["hubs"].iloc[0]))
```

```python
import re
from bs4 import BeautifulSoup

emoji_pattern = re.compile(
    "["
    "\U0001F600-\U0001F64F"
    "\U0001F300-\U0001F5FF"
    "\U0001F680-\U0001F6FF"
    "\U0001F1E0-\U0001F1FF"
    "\U00002702-\U000027B0"
    "\U000024C2-\U0001F251"
    "]+", flags=re.UNICODE
)

stop_words = {"и","в","во","не","что","он","на","я","с","со","как","а","то",
              "все","она","так","его","но","да","ты","к","у","же","вы","за",
              "бы","по","ее","мне","было","вот","от","меня","мы","о","из",
              "ему","теперь","когда","даже","ну","вдруг","ли","если","уже",
              "или","ни","быть","был","него","до","вас","нибудь","опять",
              "уж","вам","ведь","там","потом","себя","ничего","ей","может",
              "они","тут","где","есть","надо","ней","для","мы","вы",
              "the","and","is","in","to","of","a","on","this","that","it",
              "with","as","was","are"}

def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    text = str(text)
    try:
        soup = BeautifulSoup(text, "lxml")
        text = soup.get_text(separator=" ")
    except:
        pass
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = emoji_pattern.sub(r"", text)
    text = re.sub(r"\d+", " ", text)
    text = text.lower()
    tokens = re.findall(r"[a-zA-Zа-яА-ЯёЁ]+", text)
    tokens = [t for t in tokens if t not in stop_words]
    
    return " ".join(tokens)
```

```python
df["clean_text"] = df["text"].apply(preprocess_text)
df["clean_keywords"] = df["keywords"].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
```

```python
df["hubs"] = df["hubs"].apply(lambda hubs: [h.lower().strip() for h in hubs])
```

```python
df.columns
```

```python

df['keywords_str'] = df['clean_keywords'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')
```

```python
df['hubs']
```

```python
from collections import Counter

hub_counts = Counter([h for lst in df.hubs for h in lst])
valid_hubs = {h for h, c in hub_counts.items() if c >= 20}
print("Было хабов:", len(hub_counts))
print("Останется хабов:", len(valid_hubs))


df["hubs"] = df.hubs.apply(lambda lst: [h for h in lst if h in valid_hubs])
df = df[df["hubs"].apply(len) > 0]

print("Размер после фильтрации:", len(df))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df["hubs"])
```

```python
from sklearn.model_selection import train_test_split

X_text_train, X_text_test, \
X_title_train, X_title_test, \
X_keywords_train, X_keywords_test, \
y_train, y_test = train_test_split(
    df["clean_text"],
    df["title"],
    df["keywords_str"],
    y,
    test_size=0.2,
    random_state=42
)
```

```python
y
```

```python
from sklearn.feature_extraction.text import HashingVectorizer
from scipy.sparse import hstack

hv_text = HashingVectorizer(n_features=150000, ngram_range=(1,2), alternate_sign=False)
hv_title = HashingVectorizer(n_features=1000, ngram_range=(1,2), alternate_sign=False)
hv_keywords = HashingVectorizer(n_features=500, ngram_range=(1,1), alternate_sign=False)

X_text_train_vect = hv_text.transform(X_text_train)
X_title_train_vect = hv_title.transform(X_title_train)
X_keywords_train_vect = hv_keywords.transform(X_keywords_train)

X_train = hstack([X_text_train_vect, X_title_train_vect, X_keywords_train_vect])
```

```python
X_text_test_vect = hv_text.transform(X_text_test)
X_title_test_vect = hv_title.transform(X_title_test)
X_keywords_test_vect = hv_keywords.transform(X_keywords_test)

X_test = hstack([X_text_test_vect, X_title_test_vect, X_keywords_test_vect])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import jaccard_score
import joblib
import json
import numpy as np

models = {
    "MultinomialNB": OneVsRestClassifier(MultinomialNB()),
    "SGD_hinge": OneVsRestClassifier(SGDClassifier(loss="hinge", max_iter=2000)),
    "SGD_log": OneVsRestClassifier(SGDClassifier(loss="log_loss", max_iter=2000)),
    "SGD_huber": OneVsRestClassifier(SGDClassifier(loss="modified_huber", max_iter=2000)),
}

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)

    results = {
        "Jaccard micro": jaccard_score(y_test, y_pred, average='micro'),
        "Jaccard macro": jaccard_score(y_test, y_pred, average='macro')
    }

    return results, y_pred


all_results = {}

for name, model in models.items():
    print(f"Обучение {name}")
    model.fit(X_train, y_train)
    results, y_pred = evaluate(model, X_test, y_test)
    all_results[name] = results
    joblib.dump(model, f"/Users/darianaumenko/Downloads/{name}.pkl")
    with open(f"/Users/darianaumenko/Downloads/{name}_metrics.json", "w") as f:
        json.dump(results, f, indent=4)
    np.savez_compressed(f"/Users/darianaumenko/Downloads/{name}_preds.npz", y_pred=y_pred)
    print(name, results, "\n")
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

model = OneVsRestClassifier(
    LogisticRegression(
        max_iter=2000,
        solver='saga',
        n_jobs=-1
    )
)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

results = {
    "Jaccard micro": jaccard_score(y_test, y_pred, average='micro'),
    "Jaccard macro": jaccard_score(y_test, y_pred, average='macro')
}
joblib.dump(model, f"/Users/darianaumenko/Downloads/LogisticRegression.pkl")
with open(f"/Users/darianaumenko/Downloads/LogisticRegression_metrics.json", "w") as f:
    json.dump(results, f, indent=4)
np.savez_compressed(f"/Users/darianaumenko/Downloads/LogisticRegression_preds.npz", y_pred=y_pred)
```
